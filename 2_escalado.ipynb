# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

### 1. ¿Qué diferencias encuentras entre las estadísticas originales y las del escalado MinMax?


```python
print("Datos originales")
data.describe()

# El escalado MinMax transforma todas las variables para que sus valores queden dentro del intervalo [0,1]. 
# A diferencia de los datos originales, donde cada variable tiene diferentes rangos y unidades, 
# en MinMax el mínimo de cada característica es 0 y el máximo es 1. La forma de la distribución se conserva, 
# pero cambian la media y la desviación estándar.
```

```python
print("Datos escalados con MinMax")
data_minmax.describe()
```

```python
comparacion = pd.concat(
    [data.describe(), data_minmax.describe()],
    axis=1,
    keys=['Original', 'MinMax']
)

comparacion
```

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
plt.figure(figsize=(10,5))

plt.hist(
    data['sepal length (cm)'],
    bins=15,
    alpha=0.5,
    label='Original'
)

plt.hist(
    data_minmax['sepal length (cm)'],
    bins=15,
    alpha=0.5,
    label='MinMax'
)

plt.xlabel('Sepal Length')
plt.ylabel('Frecuencia')

plt.title('Distribución antes y después del escalado')

plt.legend()

plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
plt.figure(figsize=(8,6))

plt.scatter(
    data_standard['sepal length (cm)'],
    data_standard['sepal width (cm)']
)

plt.xlabel('Sepal Length escalado')

plt.ylabel('Sepal Width escalado')

plt.title('StandardScaler')

plt.show()
```

```python
# Para el conjunto de datos Iris considero que StandardScaler es la técnica más adecuada, 
# ya que centra los datos alrededor de cero y los escala utilizando la desviación estándar. 
# Esto es especialmente útil para algoritmos que dependen de distancias o cálculos estadísticos, como KNN, SVM y PCA. 
# Aunque MinMaxScaler también es una opción válida,
# StandardScaler suele ser más robusto cuando se trabaja con modelos de aprendizaje automático.
```
